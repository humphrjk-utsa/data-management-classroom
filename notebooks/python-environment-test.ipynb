## 4. Machine Learning Packages

```python
# Test machine learning packages
ml_packages = {
    'sklearn': 'Scikit-learn ML library',
    'statsmodels': 'Statistical modeling',
    'scipy.stats': 'Statistical functions'
}

print("ü§ñ Testing Machine Learning Packages:")
ml_successful = []
ml_failed = []

for package, description in ml_packages.items():
    try:
        __import__(package)
        print(f"‚úÖ {package:15} - {description}")
        ml_successful.append(package)
    except ImportError as e:
        print(f"‚ùå {package:15} - MISSING: {str(e)}")
        ml_failed.append(package)

print(f"\nüìä ML Packages Status: {len(ml_successful)}/{len(ml_packages)} successful")

# Test scikit-learn functionality
if 'sklearn' in ml_successful:
    from sklearn import datasets, linear_model, model_selection
    from sklearn.metrics import mean_squared_error, r2_score
    import sklearn
    
    print(f"\nüß† Testing Scikit-learn (version {sklearn.__version__}):")
    
    # Load sample dataset
    diabetes = datasets.load_diabetes()
    X, y = diabetes.data[:50], diabetes.target[:50]  # Use smaller subset for speed
    
    # Split data
    X_train, X_test, y_train, y_test = model_selection.train_test_split(
        X, y, test_size=0.3, random_state=42
    )
    
    # Train model
    model = linear_model.LinearRegression()
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"‚úÖ Model training: Linear regression trained on {len(X_train)} samples")
    print(f"‚úÖ Predictions: MSE = {mse:.2f}, R¬≤ = {r2:.3f}")
    print(f"‚úÖ Model coefficients shape: {model.coef_.shape}")

# Test scipy.stats
if 'scipy.stats' in ml_successful:
    from scipy import stats
    print(f"\nüìä Testing SciPy Stats:")
    
    # Generate sample data
    data1 = np.random.normal(0, 1, 100)
    data2 = np.random.normal(0.5, 1, 100)
    
    # Perform t-test
    t_stat, p_value = stats.ttest_ind(data1, data2)
    print(f"‚úÖ T-test: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
    
    # Test distribution fitting
    dist_data = np.random.normal(5, 2, 1000)
    mean_est, std_est = stats.norm.fit(dist_data)
    print(f"‚úÖ Distribution fitting: estimated mean = {mean_est:.2f}, std = {std_est:.2f}")

if ml_failed:
    print(f"‚ö†Ô∏è Missing ML packages: {', '.join(ml_failed)}")
else:
    print("\nüéâ All machine learning packages are working perfectly!")
```
